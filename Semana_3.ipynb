
#MI PROPIO ANALISIS

En una empresa se busca comprobar los siguientes datos: lineas de codigo, costo y satisfaccion del cliente.
Con base a la respuesta del analisis con 30 registros, se eligira el idioma de progamacion mas eficiente.

```python
import pandas as pd
```

```python
data=pd.read_csv('SoftwareProject_DataSet.csv')
```

```python
data.head(n=30)
```

```python
population_size = len(data)
print("El tamaño de la población es: ", population_size)
```

```python
sample_size = 60
sample = data.sample(n = sample_size, random_state=15)
```

```python
display(sample)
```

```python
sample_size = len(sample)
print(sample_size)
```

```python
sample_size = 60
interval_k = population_size // sample_size
```

```python
efficient_sample = data.iloc[::interval_k]
print("Muestra obtenida por muestreo eficiente")
display(efficient_sample)
```

```python
grouped_by_dev_language = data.groupby('Idioma de programación')
grouped_by_dev_language.head
```

```python
group_by_analysis = grouped_by_dev_language[['Líneas de código', 'Costo (USD)', 'Satisfacción del cliente']].mean()
print('Líneas de código', 'Costo (USD)', 'Satisfacción del cliente')
display(group_by_analysis)
```

Con base a los datos mostrados en los registros yo concluyo con que el idioma de programacion Kotlin es el mas eficiente para el trabajo a realizar en la empresa.
