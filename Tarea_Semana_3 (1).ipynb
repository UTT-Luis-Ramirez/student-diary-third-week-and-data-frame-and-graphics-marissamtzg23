HISTORIGRAMAS

*¿Como son? ¿Como se crean?*

Un histograma es un gráfico de barras que muestra cómo se distribuyen los datos en diferentes intervalos. Cada barra representa un rango de valores, y su altura muestra cuántos datos caen en ese rango.

Cómo se hace:
Recolecta tus datos.
Divide los datos en intervalos (por ejemplo, 0-10, 11-20, etc.).
Cuenta cuántos datos hay en cada intervalo.
Dibuja las barras: cada barra representa un intervalo y su altura es la cantidad de datos en ese intervalo.

#DIAGRAMAS DE TALLO Y HOJA

Los diagramas de tallo y hoja son una forma sencilla de representar datos numéricos, mostrando tanto la distribución como la forma de los datos. Es como una mezcla entre una tabla y un gráfico, y es útil para ver la "forma" de los datos rápidamente.

Qué son:
El tallo es la parte que representa la primera parte del número (por ejemplo, las decenas).
La hoja es la última parte del número (por ejemplo, las unidades).
Por ejemplo, con los números 23, 25, 27:

El tallo sería "2" (la decena).
Las hojas serían "3", "5", y "7" (las unidades).
Cómo se crea un diagrama de tallo y hoja:
Ordena los datos: Asegúrate de que estén en orden ascendente.
Divide los números en tallo y hoja: Para cada número, separa la parte de las decenas (tallo) y la parte de las unidades (hoja).
Dibuja una tabla: En la columna de la izquierda pones los tallos, y en la columna de la derecha, las hojas.
Agrupa las hojas: Si hay varios números con el mismo tallo, agrúpalos en la misma línea.
