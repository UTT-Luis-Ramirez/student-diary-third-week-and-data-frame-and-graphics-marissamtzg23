PROBABILIDAD Y ESTADÍSTICA

EJERCICIO DE EJEMPLO:

Un administrador de red analiza el tráfico generado por direcciones IP en una red empresarial y detecta que el 30% del tráfico proviene de una única IP. Conclusión: Se descubrió que esa IP corresponde a un ataque DDoS. Al limitar las conexiones desde esa dirección, se estabilizó el tráfico de la red.


EXPLICACIÓN DEL COMO SE PUEDE ANALIZAR EL SIGUIENTE CASO EN PROBABILIDAD Y ESTADÍSTICA

 El administrador de la red vio que el 30% del tráfico venía de una sola IP. Normalmente, el tráfico de todas las IPs debería estar más distribuido, pero este caso era anómalo.

Entonces, lo que hizo fue analizar los datos y ver si este tipo de tráfico era común o no. Al notar que esa IP estaba generando mucho más tráfico que las demás, decidió que probablemente era un ataque DDoS (un intento de sobrecargar la red con peticiones).

Con esa información, limitó las conexiones desde esa IP y, como resultado, el tráfico de la red volvió a la normalidad. Esto fue posible gracias a observar patrones en los datos y aplicar un poco de probabilidad para identificar que algo no estaba bien.

Básicamente, usó la estadística para detectar lo raro y tomar una accion. El administrador usó estadística al comparar el tráfico habitual de la red con el 30% de tráfico proveniente de una sola IP. Esto rompió el patrón normal de distribución de tráfico, lo que le dio una señal de que algo raro ocurría (probablemente un ataque DDoS).

Al identificar que esta IP generaba un volumen mucho más alto de lo esperado, aplicó una acción basada en datos (limitar conexiones de esa IP), estabilizando el tráfico de la red.

En resumen, usó estadística descriptiva para detectar la anomalía y probabilidad para deducir que era un ataque.
